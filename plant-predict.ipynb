```python
yaml_text = """
train: /content/data/train/images
val: /content/data/valid/images
test: /content/data/test/images

names:
  0: early
  1: growth
  2: harvest
  3: maturity
"""

with open('/content/data/data.yaml', 'w') as f:
    f.write(yaml_text)
```

```python
from ultralytics import YOLO

model = YOLO('yolo11n.pt')   # or yolo11s.pt for better accuracy

model.train(
    data='/content/data/data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    patience=20
)
```

```python
model.val()
```
